```python
import pymysql
import pandas as pd

def insert_food_data(file_path):
    # MySQL 연결 설정
    conn = pymysql.connect(
        host="localhost", 
        port=3306, 
        db="final_pj", 
        user="humanda", 
        passwd="1234"
    )
    
    cursor = conn.cursor()

    # CSV 파일 읽기
    data = pd.read_csv(file_path)

    # 테이블 생성
    create_table_query = """
    CREATE TABLE IF NOT EXISTS food_recipes (
        food_no INT PRIMARY KEY,
        name VARCHAR(100),
        ingredients TEXT,
        recipe TEXT,
        food_name_eng TEXT
    );
    """
    cursor.execute(create_table_query)

    # 데이터 삽입
    sql = """INSERT INTO food_recipes (food_no, name, ingredients, recipe, food_name_eng)
             VALUES (%s, %s, %s, %s, %s)
             ON DUPLICATE KEY UPDATE
                name = VALUES(name),
                ingredients = VALUES(ingredients),
                recipe = VALUES(recipe),
                food_name_eng = VALUES(food_name_eng)"""
    
    for index, row in data.iterrows():
        cursor.execute(sql, [row['food_no'], row['name'], row['ingredients'], row['recipe'], row['food_name_eng']])

    # 변경 사항 커밋
    conn.commit()

    # 리소스 정리
    cursor.close()
    conn.close()

# CSV 파일 경로
file_path = "data-files/food_final.csv"

# 함수 실행
insert_food_data(file_path)
```
