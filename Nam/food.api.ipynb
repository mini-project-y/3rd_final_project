```python
import requests
import json

# 네이버 API 키 설정
client_id = ''
client_secret = ''

# 검색 함수 정의
def search_restaurants(query):
    url = "https://openapi.naver.com/v1/search/local.json" 
    headers = {
        "X-Naver-Client-Id": client_id,
        "X-Naver-Client-Secret": client_secret,
    }
    params = {
        "query": query, 
        "display": 5,  
        "start": 1,    
        "sort": "random"
    }
    # API 요청
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        # 응답 데이터 출력
        data = response.json()
        print(json.dumps(data, indent=4, ensure_ascii=False))  # JSON 데이터를 보기 좋게 출력
        return data
    else:
        print(f"Error {response.status_code}: {response.text}")
        return None

# 테스트: 검색 키워드 입력
query = "카페"
response_data = search_restaurants(query)
```

```python
import requests
import json

# 카카오 API 키 설정
KAKAO_API_KEY = ""

# 음식점 검색 함수
def search_restaurants_nearby(query, latitude, longitude, radius=500):
    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    headers = {
        "Authorization": f"KakaoAK {KAKAO_API_KEY}"
    }
    params = {
        "query": query,  
        "x": longitude,  
        "y": latitude,   
        "radius": radius,  
        "sort": "distance", 
        "size": 15 
    }

    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        data = response.json()
        return data["documents"]  
    else:
        print(f"Error {response.status_code}: {response.text}")
        return None


latitude = 37.4979  
longitude = 127.0276 
query = "파스타"

restaurants = search_restaurants_nearby(query, latitude, longitude)

# 결과 출력
if restaurants:
    for idx, place in enumerate(restaurants, start=1):
        print(f"{idx}. {place['place_name']} - {place['address_name']} (거리: {place['distance']}m)")
else:
    print("검색 결과 없음")
```
