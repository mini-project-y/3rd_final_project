```python
train_ratio = 0.1
validation_ratio = 0.05
test_ratio = 0.05

food_name_dirs = os.listdir(os.path.join(src_base_dir, src_sub_dirs[0]))
for dir in food_name_dirs:
    files = os.listdir(os.path.join(src_base_dir, src_sub_dirs[0], dir))
    files = [file for file in files if '.jpg' in file]
    sample_cnt = len(files) * (train_ratio + validation_ratio)
    sample_cnt = int(sample_cnt)
    sample_idx = random.sample(range(0, len(files)), sample_cnt)
    split_idx = int(len(sample_idx) * 2/3)
    train_idx = sample_idx[:split_idx]
    validation_idx = sample_idx[split_idx:]
    for idx in train_idx:
        # print(os.path.join(src_base_dir, src_sub_dirs[0], dir, files[idx]))
        src_path = os.path.join(src_base_dir, src_sub_dirs[0], dir, files[idx])
        target_path = os.path.join(target_base_dir, target_sub_dirs[0], dir, files[idx])
        shutil.copy(src_path, target_path)

    for idx in validation_idx:
        # print(os.path.join(src_base_dir, src_sub_dirs[0], dir, files[idx]))
        src_path = os.path.join(src_base_dir, src_sub_dirs[0], dir, files[idx])
        target_path = os.path.join(target_base_dir, target_sub_dirs[1], dir, files[idx])
        shutil.copy(src_path, target_path)

food_name_dirs1 = os.listdir(os.path.join(src_base_dir, src_sub_dirs[1]))
for dir in food_name_dirs1:
    files = os.listdir(os.path.join(src_base_dir, src_sub_dirs[1], dir))
    files = [file for file in files if '.jpg' in file]
    sample_cnt = len(files) * test_ratio
    sample_cnt = int(sample_cnt)
    test_idx = random.sample(range(0, len(files)), sample_cnt)
    for idx in test_idx:
        src_path = os.path.join(src_base_dir, src_sub_dirs[1], dir, files[idx])
        target_path = os.path.join(target_base_dir, target_sub_dirs[2], dir, files[idx])
        shutil.copy(src_path, target_path)
```

```python
def reset_dataset_dir():
    src_base_dir = "D:/dataset"
    src_sub_dirs = ['train', 'validation']

    target_base_dir = "D:/dataset2"
    target_sub_dirs = ['train', 'validation', 'test']
    if os.path.exists(target_base_dir):
        shutil.rmtree(target_base_dir)

    os.mkdir(target_base_dir)

    food_name_dirs = os.listdir(os.path.join(src_base_dir, src_sub_dirs[0]))
    for sub_dir in target_sub_dirs:
        os.mkdir(os.path.join(target_base_dir, sub_dir))
        for food_name_dir in food_name_dirs:
            os.mkdir(os.path.join(target_base_dir, sub_dir, food_name_dir))
```

```python
def copy_files_from_src(train_ratio=0.1, validation_ratio=0.05, test_ratio = 0.1):

    food_name_dirs = os.listdir(os.path.join(src_base_dir, src_sub_dirs[0]))
    for dir in food_name_dirs:
        files = os.listdir(os.path.join(src_base_dir, src_sub_dirs[0], dir))
        files = [file for file in files if '.jpg' in file]
        sample_cnt = len(files) * (train_ratio + validation_ratio)
        sample_cnt = int(sample_cnt)
        sample_idx = random.sample(range(0, len(files)), sample_cnt)
        split_idx = int(len(sample_idx) * 2/3)
        train_idx = sample_idx[:split_idx]
        validation_idx = sample_idx[split_idx:]
        for idx in train_idx:
            # print(os.path.join(src_base_dir, src_sub_dirs[0], dir, files[idx]))
            src_path = os.path.join(src_base_dir, src_sub_dirs[0], dir, files[idx])
            target_path = os.path.join(target_base_dir, target_sub_dirs[0], dir, files[idx])
            shutil.copy(src_path, target_path)

        for idx in validation_idx:
            # print(os.path.join(src_base_dir, src_sub_dirs[0], dir, files[idx]))
            src_path = os.path.join(src_base_dir, src_sub_dirs[0], dir, files[idx])
            target_path = os.path.join(target_base_dir, target_sub_dirs[1], dir, files[idx])
            shutil.copy(src_path, target_path)

    food_name_dirs1 = os.listdir(os.path.join(src_base_dir, src_sub_dirs[1]))
    for dir in food_name_dirs1:
        files = os.listdir(os.path.join(src_base_dir, src_sub_dirs[1], dir))
        files = [file for file in files if '.jpg' in file]
        sample_cnt = len(files) * test_ratio
        sample_cnt = int(sample_cnt)
        test_idx = random.sample(range(0, len(files)), sample_cnt)
        for idx in test_idx:
            src_path = os.path.join(src_base_dir, src_sub_dirs[1], dir, files[idx])
            target_path = os.path.join(target_base_dir, target_sub_dirs[2], dir, files[idx])
            shutil.copy(src_path, target_path)
```

```python
def construct_train_files(train_ratio=0.1, validation_ratio=0.05, test_ratio=0.1):
    reset_dataset_dir()
    copy_files_from_src(train_ratio, validation_ratio, test_ratio)
```
