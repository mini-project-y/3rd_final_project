```python
import requests, json
from bs4 import BeautifulSoup

def food(name):
    url = f"https://www.10000recipe.com/recipe/list.html?q={name}"
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, "html.parser")
    food_first = soup.select('a[class = "common_sp_link"]')

    # 레시피가 없을 경우 처리
    if not food_first:
        print(f"'{name}'의 레시피를 찾을 수 없습니다.")
        return None

    food_id = food_first[0]['href'].split("/")[-1]
    food_link = f"https://www.10000recipe.com/recipe/{food_id}"
    food_response = requests.get(food_link)
    food_html = food_response.text
    food_soup = BeautifulSoup(food_html, "html.parser")
    food_info = food_soup.select('script[type="application/ld+json"]')

    # JSON 데이터가 없을 경우 처리
    if not food_info:
        print(f"'{name}'의 상세 레시피를 찾을 수 없습니다.")
        return None

    result = json.loads(food_info[0].text)

    # 'recipeIngredient'와 'recipeInstructions'가 없을 경우 처리
    if 'recipeIngredient' not in result or 'recipeInstructions' not in result:
        print(f"'{name}'의 레시피 데이터가 누락되었습니다.")
        return None

    ingredient = ','.join(result['recipeIngredient'])
    recipe = [row['text'] for row in result['recipeInstructions']]
    numbered_recipe = [f"{i+1}. {step}" for i, step in enumerate(recipe)]

    fd = {
        'name': name,
        'ingredients': ingredient,
        'recipe': numbered_recipe
    }
    return fd

# 음식 리스트 처리
food_list = ["곱창구이", "꿀떡", "조개구이", "약과", "한과"]

for food_item in food_list:
    food_data = food(food_item)  # food 함수 호출
    if food_data:  # 유효한 데이터만 출력
        print(f"'{food_data['name']}'의 레시피:")
        print("재료:", food_data['ingredients'])
        print("조리법:")
        for step in food_data['recipe']:
            print(step)
        print("-" * 30)
```
